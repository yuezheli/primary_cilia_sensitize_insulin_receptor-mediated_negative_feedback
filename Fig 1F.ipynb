```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



## define functions for calculation
def Normallization(trace, preframe):
    ts = np.array(trace)
    return ts/np.min(ts[:preframe])


time_ins = np.array(range(241)) # time in seconds
```

# Fig 1G

```python
medium = pd.read_excel('Medium, BMS536924.xlsx', sheet_name = '+medium', header = 1)


plt.plot(time_ins, Normallization(medium.iloc[:, 13], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity', fontsize = 16)
plt.bar(np.array(range(121, 242, 1)), height = 0.1, width = 1, bottom = 1.78, color = 'k')
plt.text(105, 1.7, '+imaging medium', color = 'k', fontsize = 14)
plt.xlim([0, 241])
plt.ylim(1, 1.8)
plt.xlabel('time (s)', fontsize = 16)
```
