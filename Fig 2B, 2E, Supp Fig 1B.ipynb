```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



## define functions for calculation
def Normallization(trace, preframe):
    ts = np.array(trace)
    return ts/np.min(ts[:preframe])


time_ins = np.array(range(241)) # time in seconds
```

# Fig 2B

```python
# plot +bovins insulin, +bovins insulin (DMSO), +bovinen insulin (LY294002)

bins = pd.read_excel('bINS_hINS_hIGF1_175nM.xlsx', sheet_name = 'bINS_175nM', header = 1) 
bins_dmso = pd.read_excel('bovine_insulin.xlsx', sheet_name = 'DMSO', header = 1)
bins_ly = pd.read_excel('bovine_insulin.xlsx', sheet_name = 'LY294002', header = 1)

plt.figure(figsize = (6,6))
plt.subplot(311)
plt.plot(time_ins, Normallization(bins.iloc[:, 5], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(121, 242, 1)), height = 0.1, width = 1, bottom = 5, color = 'k')
plt.text(121, 4.5, '+Bovine insulin (175nM)', color = 'k')
plt.xlim([0, 241])
plt.ylim(1, 5.1)
plt.subplot(312)
plt.plot(time_ins, Normallization(bins_dmso.iloc[:, 17], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(0, 242, 1)), height = 0.2, width = 1, bottom = 4.95, color = 'r')
plt.text(0, 4.5, '+ DMSO', color = 'r')
plt.bar(np.array(range(121, 242, 1)), height = 0.1, width = 1, bottom = 4.8, color = 'k')
plt.text(121, 4.4, '+Bovine insulin (175nM)', color = 'k')
plt.xlim([0, 241])
plt.ylim(1, 5.1)
plt.subplot(313)
plt.plot(time_ins, Normallization(bins_ly.iloc[:, 25], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(0, 242, 1)), height = 0.2, width = 1, bottom = 4.95, color = 'r')
plt.text(0, 4.5, '+ LY294002', color = 'r')
plt.bar(np.array(range(121, 242, 1)), height = 0.1, width = 1, bottom = 4.8, color = 'k')
plt.text(121, 4.4, '+Bovine insulin (175nM)', color = 'k')
plt.xlim([0, 241])
plt.ylim(1, 5.1)
plt.xlabel('time (s)', fontsize = 14)
```
