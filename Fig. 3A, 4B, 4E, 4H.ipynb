```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



## define functions for calculation
def Normallization(trace, preframe):
    ts = np.array(trace)
    return ts/np.min(ts[:preframe])


time_glu = np.array(range(721)) - 121 # time in seconds, time 0 as the stimulation is added
```

# Fig 3A

```python
glu_nocilia = pd.read_excel('12min_response.xlsx', sheet_name = '-cilium, +glucose', header = 1)
gluins_nocilia = pd.read_excel('12min_response.xlsx', sheet_name = '-cilium, +glucose & insulin', header = 1)
glu_dmso12 = pd.read_excel('12min_response.xlsx', sheet_name = 'DMSO', header = 1)
glu_bms12_1 = pd.read_excel('12min_response.xlsx', sheet_name = 'BMS', header = 1)
glu_bms12_2 = pd.read_excel('12min_response.xlsx', sheet_name = '+cilium, +glucose & bms', header = 1)
glu_bms12_3 = pd.read_excel('12min_response.xlsx', sheet_name = '-cilium, +glucose & bms', header = 1)
glu_bms12 = pd.concat([glu_bms12_1, glu_bms12_2, glu_bms12_3], axis = 1)
plt.figure(figsize = (4,10))

plt.subplot(411)
plt.plot(time_glu, Normallization(glu_nocilia.iloc[:, 10], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(0, 601, 1)), height = 0.1, width = 1, bottom = 5, color = 'k')
plt.text(0, 4.5, '+Glucose', color = 'k')
plt.xlim([-121, 600])
plt.ylim(0.5, 5.1)

plt.subplot(412)
plt.plot(time_glu, Normallization(gluins_nocilia.iloc[:, 5], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(0, 601, 1)), height = 0.1, width = 1, bottom = 5, color = 'k')
plt.text(0, 4.5, '+Glucose & Ins', color = 'k')
plt.xlim([-121, 600])
plt.ylim(0.5, 5.1)

plt.subplot(413)
plt.plot(time_glu, Normallization(glu_dmso12.iloc[:, 32], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(-121, 601, 1)), height = 0.2, width = 1, bottom = 5.95, color = 'g')
plt.text(-121, 5.2, '+ DMSO', color = 'g')
plt.bar(np.array(range(0, 601, 1)), height = 0.1, width = 1, bottom = 5.8, color = 'k')
plt.text(133, 5.3, '+Glucose (25mM)', color = 'k')
plt.xlim([-121, 600])
plt.ylim(0.5, 6.1)

plt.subplot(414)
plt.plot(time_glu, Normallization(glu_bms12.iloc[:, 31], 120), color = 'gray')
plt.ylabel('normalized \n fluorescence \n intensity')
plt.bar(np.array(range(-121, 601, 1)), height = 0.2, width = 1, bottom = 5.95, color = 'r')
plt.text(-121, 5.2, '+ BMS536924', color = 'r')
plt.bar(np.array(range(0, 601, 1)), height = 0.1, width = 1, bottom = 5.75, color = 'k')
plt.text(133, 5.3, '+Glucose (25mM)', color = 'k')
plt.xlim([-121, 600])
plt.ylim(0.5, 6.1)
plt.xlabel('time (s)', fontsize = 14)
plt.show()
```

# Fig 4B

```python
# add glucose, compare with and without primary cilium

glu_ciliated = pd.read_excel('12min_response.xlsx', sheet_name = '+cilium, +glucose', header = 1)

glu_nocilia = pd.read_excel('12min_response.xlsx', sheet_name = '-cilium, +glucose', header = 1)

plt.figure(figsize = (8,3))

plt.plot(time_glu, Normallization(glu_nocilia.iloc[:, 10], 120), color = 'lightgray', label = '-cilium', linestyle = '-')
plt.plot(time_glu, Normallization(glu_ciliated.iloc[:, 0], 120), color = 'k', label = '+cilium', linestyle = ':')
plt.legend(fontsize = 14, loc = 'upper right')
plt.ylabel('normalized \n fluorescence \n intensity', fontsize = 14)
plt.bar(np.array(range(0, 601, 1)), height = 0.1, width = 1, bottom = 5, color = 'k')
plt.text(0, 4.5, '+Glucose', color = 'k', fontsize = 14)
plt.xlim([-121, 600])
plt.ylim(0.5, 5.1)
plt.xlabel('time (s)', fontsize = 14)
```
